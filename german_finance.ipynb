```python
# LISTA SALARIOS

csv_list = [['3044', '3744', '6'], ['3821', '5226', '24']] # resident start, resident end, years | attending start, attending end, years # NETTO

```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import numpy as np

def plot_yearly_graph(lista, yaxis):
    time = np.arange(len(lista))

    
    plt.plot(time, lista, marker='o')

    plt.xlabel('Anos')
    plt.ylabel(yaxis)
    
    #plt.ticklabel_format(style='plain', axis='y', scilimits=(0,0))
    plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: '{:,.0f}'.format(x)))
    
    plt.show()

yearly_salary_list = []


for i in range(len(csv_list)):
    current_yearly_increase = ((float(csv_list[i][1])/float(csv_list[i][0]))-1)/(float(csv_list[i][2])-1)
    start_monthly_salary = float(csv_list[i][0])

    for j in range(int(csv_list[i][2])):
        monthly_salary = start_monthly_salary + (start_monthly_salary*current_yearly_increase*j)
        yearly_salary = monthly_salary*12
        yearly_salary_list.append(round(yearly_salary, 2))

        if j == (int(csv_list[i][2])-1):
            if j > 15:
                for i in range(10):
                    yearly_salary_list.append(round(yearly_salary, 2))
```

```python
# VARIABLES:
STARTING_CAPITAL = 2000
SAVED_SALARY = 0.35
BOND_REAL_INTERST = 1
WORLD_EQUITIES_REAL_INTEREST = 1.05

WITHDRAW_TIMES = []
WITHDRAW_AMOUNTS = []

RISK_PROFILE = 0.98 # PERCENTAGE OF EQUITIES
```

```python
GRAPH_CUTOFF_YEARS = 35

def accumulated_capital():
    capital_list = []
    
    current_capital = STARTING_CAPITAL

    capital_list.append(current_capital)
    
    for i in range(GRAPH_CUTOFF_YEARS):
        current_capital = current_capital + (SAVED_SALARY*yearly_salary_list[i])
        current_capital = (current_capital*RISK_PROFILE*WORLD_EQUITIES_REAL_INTEREST) + (current_capital*(1-RISK_PROFILE)*BOND_REAL_INTERST)
        capital_list.append(current_capital)
    
    plot_yearly_graph(capital_list,'Capital Total (€)')
    
accumulated_capital()
```
